```python
store = pd.read_excel(r"D:\UV TECH\Python\Practice\Coffee Shop\Location Details.xlsx")
```

```python
product = pd.read_excel(r"D:\UV TECH\Python\Practice\Coffee Shop\Product details.xlsx")
product
```

```python
product1 = product.rename(columns={"Product_id":"product_id"})
```

```python
cs1 = pd.merge(coffeeshop,store, how="inner",on="store_id")
cs1
```

```python
cs2 = pd.merge(cs1,product1,how="inner",on="product_id")
```

```python
cs2
```

```python
cs2 = cs2.drop(columns=["product_category_y","product_type_y"])
```

```python
cs2
```

```python
cs2["Total_Bill"] = cs2["transaction_qty"] * cs2["unit_price"]
```

```python
cs2["Profit"] = cs2["Total_Bill"] - (cs2["Unit_Making_Price"]*cs2["transaction_qty"])
cs2
```

```python
cs2.info()
```

```python
cs2["Month_Name"] = cs2["transaction_date"].dt.month_name()
cs2
```

```python
KN = cs2[cs2["store_location"] == "Karave Nagar"]
```

```python
SN = cs2[cs2["store_location"] == "Shivaji Nagar"]
```

```python
kn_data = KN.groupby(by="Month_Name")["unit_price"].sum()
kn_data = kn_data.reindex(['January','February','March','April','May','June','July','August','September','October','November','December'])
kn_data
```

```python
sn_data = SN.groupby(by="Month_Name")["unit_price"].sum()
sn_data = sn_data.reindex(['January','February','March','April','May','June','July','August','September','October','November','December'])
sn_data
```

```python
kn_data.plot(kind="line",label = "Karave Nagar")
sn_data.plot(kind="line",label = "Shiavji Nagar")

plt.legend()
plt.show()
```
